```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Central 100 Data Analysis

Train and dev data: 100% Central catalan

Test data (analyzed here): 20% each Central, Valencià, Nord, Nord-Occidental, and Balear

```python
df=pd.read_csv("../central100_53m_v02/eval/full_eval.tsv",delimiter='\t')
```

```python
df.head()
```

```python
#Going on this example: https://seaborn.pydata.org/examples/horizontal_boxplot.html

sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 6))

sns.boxplot(
    df, x="Word Error Rate", y="Dialect", hue="Dialect",
    whis=[0, 100], width=.6, palette="vlag"
)

#sns.stripplot(df, x="Word Error Rate", y="Dialect", size=4, color=".3")

ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

```python
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(7, 6))

sns.boxplot(
    df, x="Character Error Rate", y="Dialect", hue="Dialect",
    whis=[0, 100], width=.6, palette="vlag"
)

#sns.stripplot(df, x="Word Error Rate", y="Dialect", size=4, color=".3")

ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

```python
#get all the average WER/CER
print("Dialect\tWER\tCER")
for dial in ['valencià','central','nord','balear','nord-occidental']:
    df_dial=df[df['Dialect']==dial]
    print(f"{dial}\t{df_dial['Word Error Rate'].mean()}\t{df_dial['Character Error Rate'].mean()}")
```

Seems like it would be nice to overlay all the distributions over each other in a ridgeline plot

```python
df["Model"] = "100% Central"
df_80["Model"] = "80% Central"
df_20["Model"] = "20% Central"
```

```python
combined_df = pd.concat([df,df_80,df_20])
```

```python
combined_df
```

```python
import plotly.express as px
```

```python
fig = px.violin(combined_df,y="Dialect",x="Word Error Rate",color="Model",points=False,
          violinmode="overlay",orientation="h",width=650,height=850,
          color_discrete_sequence=px.colors.qualitative.Dark24,labels={'balear':'Balear',
                                                                       'central':'Central',
                                                                       'nord':'Nord',
                                                                       'nord-occidental':'Nord-Occidental',
                                                                       'valencià':'Valencià'}).update_traces(side="positive", width=2)
fig.update_xaxes(range = [0,1])
fig.update_yaxes(labelalias={'balear':'Balear',
                            'central':'Central',
                            'nord':'Nord',
                            'nord-occidental':'Nord-Occidental',
                            'valencià':'Valencià'},
                categoryorder='category descending')
fig.update_layout(plot_bgcolor="rgb(245, 245, 245)")

fig.show()
```
